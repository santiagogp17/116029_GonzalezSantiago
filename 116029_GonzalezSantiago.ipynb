```python
#1. VALOR ABSOLUTO
num=float(input('Ingrese el número: '))#numero ingresado en flotante
comp=float(0.0)#para comparar si es mayor, menor o igual a cero
if num<comp:
    vabs=float(num*(-1.0))
if num>0:
    vabs=float(num)
if num==comp:
    vabs=float(0)
print('El valor absoluto es: ',vabs)
```

```python
#3. LEER UN NUMERO DE TRES DIGITOS Y DETERMINAR SI CONTIENE EL DIGITO 8
entero=int(input('Ingrese un numero de 3 digitos: ')) 
numabs=abs(entero)#valor absoluto del numero ingresado
num=str(numabs)#volver el numero cadena de texto
digitos=len(num)#cantidad de digitos
if digitos==3:#descide si el digito es de tres digitos o no
    for i in range(0,digitos):#analiza digitopor digito
        if num[i]=='8':#analiza si hay algun digito que sea 8
            print('Si contiene el digito 8')
            break
        if i==digitos-1 and num[digitos-1]!='8':
            print('No contiene el numero 8')
else:
    print('El numero no es de 3 digitos')
```

```python
#4. Determinar la suma de digitos pares
num=input('Ingrese un numero entero: ')
suma=0
for i in range(len(num)):#se va digito por digito
    entero=int(num[i])#vuelve el dijito de str a entero
    if (entero//2)*2==entero:#determina si el digito i es par
        suma=suma+int(num[i])#si es par lo suma
print('La suma de los digitos pares es',suma)
```

```python
#5. Leer un numero entero y determinar cuantos cincos consecutivos hay
num=input('Ingrese un numero entero: ')
if num[0]=='5' and num[1]=='5':#determina si los primeros dos digitos son 5
    cont=2
else:
    cont=0
for i in range(1,len(num)-1):#recorre desde el digito 2 al ultimo
    if num[i]=='5' and num[i+1]=='5':
        cont=cont+1
    if num[i]=='5' and num[i-1]!='5' and num[i+1]=='5':
        cont=cont+1
print(f'Tiene {cont} cincos consecutivos')
```

```python
#6. determinar si la suma de los digitos es un fibonacci
#suma de los digitos
num=input('Ingrese un numero entero:')
suma=0 #guarda la suma de los digitos
for i in range(len(num)): #recorre cada digito
    suma=suma+int(num[i]) #suma cada digito
#fibonacci
cont=0
b=1
fibonacci1=[]#guarda una parte de los numeros fibonacci
fibonacci2=[]#guarda otra parte de los numeros fibonacci
fibonacci=[]#guarda todos los numeros fibonacci
while b<int(num) and cont<int(num):#pone limite a la lista
    cont+=b 
    b+=cont
    fibonacci1.append(b)
    fibonacci2.append(cont)
    if b<=int(num) or cont<=int(num):#limita que la lista llegue hasta num
        fibonacci=sorted(fibonacci1+fibonacci2)#organiza la lista en orden
if suma in fibonacci or suma==0:#determina si la suma es un fibonacci
    print(f'La suma {suma} pertenece a la secuencia Fibonacci')
else:
    print(f'La suma {suma} no pertenece a la secuencia Fibonacci')
```

```python
#7. Leer un numero entero positivo y determinar si existen numeros repetidos
num=input('Ingrese un numero entero positivo: ')
a=''#donde se almacenaran los repetidos
for i in range(10):
    if num.count(str(i))>1:#este es count de cadenas no de listas
        a=a+str(i)+'-'#agrega el elemento en caso de que se repita
    else:
        a=a
if a!='':
    print('Numeros repetidos: ',a)
else:
    print('No hay numeros repetidos')
```

```python
#8. Multiplos de 3 y de 4 en los primeros 100 enteros
cont=0
lista3=[]#almacena los multiplos de 3
lista4=[]#almacena los multiplos de 4
print('Multiplos de 3: ')
for i in range(101):
    if (i//3)*3==i and i!=0:#determina si i es multiplo de 3
        lista3.append(i)#agrega el elemento a la lista si es multiplo
        cont+=1
        if cont==15:#limita la lista hasta los primeros 15 multiplos
            break
print(lista3)
print('Multiplos de 4:')
for s in range(i,101):
    if (s//4)*4==s:#determina si i es multiplo de 4
        lista4.append(s)#agrega el elemento a la lista si es multiplo
print(lista4)
```

```python
#9. Ingresar dos circulos y un punto con coordenadas y redios y determinar si:
#el punto esta dentro del circulo 1, dentro del circulo 2, dentro de los dos
#circulos o en ningun circulo
x1=float(input('Ingrese la coordenada X del circulo 1: '))
y1=float(input('Ingrese la coordenada Y del circulo 1: '))
r1=float(input('Ingrese el radio del circulo 1: '))
x2=float(input('Ingrese la coordenada X del circulo 2: '))
y2=float(input('Ingrese la coordenada Y del circulo 2: '))
r2=float(input('Ingrese el radio del circulo 2: '))
a=float(input('Ingrese la coordenada en X del punto: '))
b=float(input('Ingrese la coordenada en Y del punto: '))
print('RESPUESTA')
if r1==0:#el circulo no existiria
    print('El radio 1 debe ser mayor a 0')
if r2==0:#el circulo no existiria
    print('El radio 2 debe ser mayor a 0')
if r1!=0 and r2!=0:
    dist1=((a-x1)**2+(b-y1)**2)**0.5 #Distancia del punto al centro 1
    dist2=((a-x2)**2+(b-y2)**2)**0.5 #Distancia del punto al centro 2
    if dist1<=r1 and dist2<=r2:#cuando esta dentro de ambos
        print('El punto esta dentro de ambos circulos')
    if dist1<=r1 and dist2>r2:#cuando esta dentro del1
        print('El punto esta dentro del circulo 1')
    if dist1>r1 and dist2<=r2:#cuando esta dentro del 2
        print('El punto esta dentro del circulo 2')
    if dist1>r1 and dist2>r2:#cuando no esta en ninguno
        print('El punto no esta contenido en ningun circulo')
```

```python
#10. Sin utilizar el método .upper() para cadenas, escribir un programa que lea
# una cadena de texto y la imprima en mayúscula. Tenga en cuenta que el
# programa debe aceptar las letras con tilde, la ñ y la ü.
cadena=input('Inserte la cadena que desea modificar: ')
print('Cadena modificada:')
for i in range(len(cadena)):#recorre cada elemento de la cadena
    if cadena[i].islower():
        mayuscula=chr(ord(cadena[i])-32)#cambia el unicode si es minuscula
    elif cadena.isupper():
        mayuscula=cadena[i]#si es mayuscula la deja como esta
    print(mayuscula,end='')
```

```python
#12. Hacer un programa que lea dos palabras (podrían estar en mayúscula o 
#minúscula) y determine cual está primero en el diccionario. El programa debe 
#soportar letras con tilde, la ü y la ñ.

#definicion de variables
palabra1=input('ingrese la primer palabra: ')
palabra2=input('ingrese la segunda palabra: ')
palabra1=palabra1.lower()
palabra2=palabra2.lower()
#define el rango para los ciclos
if len(palabra1) < len(palabra2):
    rango=len(palabra2)   
if len(palabra1) >= len(palabra2):
    rango=len(palabra1)
especiales=('áéíóúüñ')
normales=('aeiouun')
#ciclos para soportar las letras diferentes
for i in range (len(especiales)):
    for j in range (len(palabra1)):
         if especiales[i] == palabra1[j]:
             palabra1 = palabra1.replace(especiales[i],normales[i])
for i in range (len(especiales)):
    for j in range (len(palabra2)):
         if especiales[i] == palabra2[j]:
             palabra2 = palabra2.replace(palabra2[j],normales[i])
#determinacion del orden de las palabras
if palabra1<palabra2:
    print(f'Va primero la palabra {palabra1} en el diccionario')
if palabra1>palabra2:
    print(f'Va primero la palabra {palabra2} en el diccionario')
if palabra1==palabra2:
    print(f'Son las mismas palabras')
```

```python
#13. Escribir un programa que lea una cadena de texto y la imprima como un
#triángulo.
# Se definen las funciones a utilizar:
def leer_texto(enunciado):
    while True:
        try:
            s = input(enunciado)
            break
        except (ValueError, KeyboardInterrupt, EOFError):
            print('Por favor ingrese un un texto. Intente nuevamente.')
    return s
# Programa principal:
cadena = leer_texto('Introduzca una cadena de texto: ')
# Se le quitan todos los espacios a la cadena
cad_sin_esp = cadena.replace(' ', '')
longitud = len(cad_sin_esp)
num_car_renglon = 1 # No. de caracteres de la base del triángulo.
cont_renglones = 0
while True:
    longitud -= num_car_renglon
    if longitud <= 0:
        break
    cont_renglones += 1
    num_car_renglon += 2
# En este punto, la variable longitud habrá cambiado su valor, así que se le
# vuelve a asignar el que tenía inicialmente.
longitud = len(cad_sin_esp)
num_car_renglon -= 2  # Tras el último ciclo esta variable quedó con un exce-
# dente de 2. Se compensa restándoselo.
start = 0
paso = 1
for i in range(0, cont_renglones):
    print(cad_sin_esp[start:start+paso].center(num_car_renglon))
    start += paso  # El inicio del slicing para el siguiente renglón se incre-
    # menta según la variable "paso"
    paso += 2  # Y para cada rengón se imprimen 2 caracteres más, por esto, se
    # aumenta la variable paso de 2 en 2.
# Para imprimir la última línea:
suma = 0  # Cuántos caracteres se han podido imprimir en el triángulo.
for i in range(1, num_car_renglon+1, 2):
    suma += i
if suma < longitud:  # Verifica si quedaron caracteres sobrantes por fuera del
                     # triángulo.
    print(cad_sin_esp[suma:longitud].center(num_car_renglon))
```

```python
#14. Leer una cadena de texto y organice alfabéticamente cada una de las letras
#que la componen, repitiendo cada una tantas veces como se encuentra.
cad=input('Inserte la cadena que desea organizar alfabeticamente: ')#cadena
cad=cad.lower()#cadena ingresada en minusculas
alfabetica=[]#lista a la que se agregara el codigo unicode de cada elemento
ordenalfa=[]#lista donde se pondran en orden los elementos segun su codigo
#ciclo que crea una lista donde los elementos son el orden de cada letra
for i in cad:
    alfabetica.append(ord(i))
ordennum=sorted(alfabetica)#se organiza la lista en orden
for i in ordennum:
    ordenalfa.append(chr(i))#en el orden que estan, cada elemento pasa a str
print('La cadena ordenada alfabeticamente es: ',''.join(ordenalfa),end='')
```
